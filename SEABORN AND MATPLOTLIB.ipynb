```python
iris =sns.load_dataset('iris')
```

```python

#allow us to plot pairwise relationships between variables within a dataset
sns.pairplot(iris)
```

```python
#specific column relationship
sns.pairplot(iris,hue='species')
```

```python
#specific column relationship
sns.pairplot(iris,hue='species' ,diag_kind='hist')
```

```python
#specific column relationship
sns.pairplot(iris,kind='kde')
```

```python
g=sns.relplot(kind='scatter',data=iris, x='sepal_width', y='petal_width')
g.set(title="IRIS_DATA")
```

```python
g=sns.relplot(kind='scatter',data=iris, x='sepal_width', y='petal_width', hue='species')
g.set(title="IRIS_DATA")
```
